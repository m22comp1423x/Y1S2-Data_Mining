### a) Write python codes that can detect the faces and show the faces clearly by making rectangle frame boarding the face of your colors. (15 marks)

```python
# imports
import cv2

from IPython.display import display, clear_output

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
Cristiano = cv2.imread('Cristiano_Ronaldo_2018.jpg')
babbon = cv2.imread('baboon.png')
```

```python
fig, ax = plt.subplots(1, 2)
clear_output()
ax[0].imshow(cv2.cvtColor(Cristiano, cv2.COLOR_BGR2GRAY), cmap='gray')
ax[0].axis('off')
ax[0].set_title('Cristiano')
ax[1].imshow(cv2.cvtColor(babbon, cv2.COLOR_BGR2GRAY), cmap='gray')
ax[1].axis('off')
ax[1].set_title('Baboon')
```

```python
cascadePath = "haarcascade_frontalface_default.xml"
# initialize a model for detecting whether a certain portion of an image contains a face
faceCascade = cv2.CascadeClassifier(cascadePath)
```

```python
gray_c = cv2.cvtColor(Cristiano, cv2.COLOR_BGR2GRAY)
```

```python
faces = faceCascade.detectMultiScale(
        gray_c,
        scaleFactor=1.1,
        minNeighbors=6,
        minSize=(100, 100),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
# output of the classifier
faces
```

```python
from matplotlib.patches import Rectangle

# visualize the detected face
fig, ax = plt.subplots()
ax.imshow(gray_c, cmap='gray')

# iterate over all the detected faces
for face in faces:
    # retrieve the coordinates of the position of the current face, and its size    
    (x_c, y_c, w_c, h_c) = face
    # draw a rectangle where the face is detected    
    ax.add_artist(Rectangle((x_c, y_c), w_c, h_c, fill=False, lw=3, color='green'))

ax.axis('off')
```

```python
gray_b = cv2.cvtColor(babbon, cv2.COLOR_BGR2GRAY)

# openCv algorithm to scan the image looking for haar features
# faces = ...
# SOLUTION:
faces = faceCascade.detectMultiScale(
        gray_b,
        scaleFactor=1.1,
        minNeighbors=6,
        minSize=(100, 100),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
# output of the classifier
faces
```

```python
fig, ax = plt.subplots()
ax.imshow(gray_b, cmap='gray')

# iterate over all the detected faces
for face in faces:
    # retrieve the coordinates of the position of the current face, and its size    
    (x_b, y_b, w_b, h_b) = face
    # draw a rectangle where the face is detected    
    ax.add_artist(Rectangle((x_b, y_b), w_b, h_b, fill=False, lw=3, color='green'))

ax.axis('off')
```
